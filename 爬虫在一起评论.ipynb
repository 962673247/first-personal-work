```python
import requests
import re
import requests
import lxml.html
import csv
from selenium import webdriver
import time
import json
import emoji
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from datetime import datetime
headers={'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.183 Mobile Safari/537.36'}
firsturl="https://v.qq.com/x/cover/mzc00200jg5gfcq/u0034zxdhdi.html"
```

```python
res=requests.get(firsturl,headers=headers)
selector=lxml.html.fromstring(res.text)
nextid=selector.xpath("//*[@id='video_scroll_wrap']/div[4]/div[1]/span")
idlist=list()
for id in nextid:
    idlist.append(id.xpath("@id")[0])#获取每一集的网址 然后循环构造
print(idlist)
```

```python
numberlist=list()
for id in idlist:
    bigurl="https://v.qq.com/x/cover/mzc00200jg5gfcq/"+id+".html"
    res=requests.get(bigurl,headers=headers)
    number=re.findall('"comment_id":"(.*?)"',res.text)[0]#通过上面每一集的网址获取每一集的抓包参数
    numberlist.append(number)
    print(number)
```

```python
print(numberlist)#检测数量是否正确 根据这些数字来构造url获取评论数
```

```python
with open('211806150pinglun.txt','a',newline='',encoding="utf-8") as file:
    for x in range(len(numberlist)):
        number=numberlist[x]
        newdata=int(time.time()*1000)
        hasnext="true"
        lastnumber="0"#初始值为0 后面根据json数据去生成
        j=0
        while(hasnext=="true"):
            jsonurl="https://video.coral.qq.com/varticle/"+number+"/comment/v2?callback=_varticle"+number+"commentv2&orinum=10&oriorder=o&pageflag=1&cursor="+lastnumber+"&scorecursor=0&orirepnum=2&reporder=o&reppageflag=1&source=132&_="+str(newdata)
            res=requests.get(jsonurl,headers=headers)
            res.encoding="utf-8"
            textlist=re.findall('"content":"(.*?)"',res.text)
            hasnext=re.findall('"hasnext":(.*?),',res.text)[0]
            lastnumber=re.findall('"last":"(.*?)"',res.text)[0]
            newdata=newdata+1
            i=0
            for text in textlist:
                file.writelines(text)
                i=i+1
                j=j+1
                print("正在爬取第"+str(x+1)+"集评论第"+str(j)+"条成功")
                if(i==10):
                    break
```
